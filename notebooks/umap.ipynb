# Methode UMAP :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import umap
from pathlib import Path
```

## Chargement du dataset et affichage des premières lignes :

```python
data_path = Path("../data/city_lifestyle_dataset.csv")

df = pd.read_csv(data_path)

df.head()
```

```python
X = df.select_dtypes(include=[np.number]).to_numpy()
print("Shape de X :", X.shape)
```

```python
reducer = umap.UMAP(n_components=2, random_state=42)
X_2d = reducer.fit_transform(X)

print("Shape projection :", X_2d.shape)
```

## Projection des données en dimension 2 :

```python
plt.figure(figsize=(7,5))
plt.scatter(X_2d[:, 0], X_2d[:, 1], s=10)
plt.title("Projection UMAP (2D)")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.show()
```
